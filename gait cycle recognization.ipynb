```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('test1.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
R_value = data.iloc[:,0]
L_value = data.iloc[:,1]
print(R_value)
print(L_value)
L_value[27]
```

```python
j=0
c=0
k=0
R_feature=[]
for i in range(len(R_value)):
    if R_value[i] == 0 and R_value[i+1] != 0:
        print("period start:  " ,i)
        c=c+1
        if c%2 !=0:
            j=i
            R_period=R_value[k:j]
            R_mean=np.mean(R_period)
            R_std=np.std(R_period)
            R_max=np.max(R_period)
            R_ratio=R_max/R_mean
            R_time=j-k
            feature=[R_mean,R_std,R_max,R_ratio,R_time,k,j]
            R_feature.append(feature)
            print("period:",R_period)
        if c %2 ==0:
            k=i
            R_period=R_value[j:k]
            R_mean=np.mean(R_period)
            R_std=np.std(R_period)
            R_max=np.max(R_period)
            R_ratio=R_max/R_mean
            R_time=k-j
            feature=[R_mean,R_std,R_max,R_ratio,R_time,j,k]
            R_feature.append(feature)
            print("period:",R_period)
print(R_feature[1:])
```

```python
t=pd.DataFrame(R_feature[1:4])
print(t)
R_bvaluemean=np.mean(t.iloc[0:2,0])
R_bstdmean=np.mean(t.iloc[0:2,1])
R_bmaxmean=np.mean(t.iloc[0:2,2])
R_bratiomean=np.mean(t.iloc[0:2,3])
R_bperiodmean=np.mean(t.iloc[0:2,4])
R_analysis=[]
b_analysis=[R_bvaluemean,R_bstdmean,R_bmaxmean,R_bratiomean,R_bperiodmean]
R_analysis.append(b_analysis)
print("beginning period:",R_analysis)

q=pd.DataFrame(R_feature[5:15])
print(q)
R_svaluemean=np.mean(t.iloc[0:9,0])
R_sstdmean=np.mean(t.iloc[0:9,1])
R_smaxmean=np.mean(t.iloc[0:9,2])
R_sratiomean=np.mean(t.iloc[0:9,3])
R_speriodmean=np.mean(t.iloc[0:9,4])
s_analysis=[R_svaluemean,R_sstdmean,R_smaxmean,R_sratiomean,R_speriodmean]
R_analysis.append(s_analysis)
print("beginning and stable period: ",R_analysis)
```

```python
R_ratio=pd.DataFrame(R_analysis)
print(R_ratio)
R_bsvalue=R_ratio.iloc[0,0]/R_ratio.iloc[1,0]
R_bsstd=R_ratio.iloc[0,1]/R_ratio.iloc[1,1]
R_bsmax=R_ratio.iloc[0,2]/R_ratio.iloc[1,2]
R_bsratio=R_ratio.iloc[0,3]/R_ratio.iloc[1,3]
R_bsperiod=R_ratio.iloc[0,4]/R_ratio.iloc[1,4]
R_bs=[R_bsvalue,R_bsstd,R_bsmax,R_bsratio,R_bsperiod]
R_bs=pd.DataFrame(R_bs,index=['R_bsmean','R_bsstd','R_bsmax','R_bsratio','R_bsperiod'])
print(R_bs.T)
```

```python
df1=pd.DataFrame(R_feature[1:],columns=["R_mean",'R_std','R_max','R_ratio','R_period','R_start','R_end'])
df2=pd.DataFrame(R_bs.T)
print(df2)
R_file = pd.concat([df1,df2],axis=1)
R_file = pd.DataFrame(R_file)
print(R_file)
```

```python
R_file.to_csv('resultR.csv',columns=["R_mean",'R_std','R_max','R_ratio','R_period','R_start','R_end','R_bsmean','R_bsstd','R_bsmax','R_bsratio','R_bsperiod'],index=False,sep=',')
```
